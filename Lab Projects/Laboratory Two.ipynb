---------

In order to get reasonable results from our qasm simulator we need to add a measurement to our circuit.<br>
This is done with the measure( ) procedure. This procedure takes two parameters.<br>
The first parameter is the list of qubits that will be measured, and the second is a list of the classical
bits where the result will be placed. <br>
We also need to change our QuantumCircuit call to add 2 classical bits to our circuit.

```python
qc = QuantumCircuit(2,2)#The second parameter is the classical bit, so we added 2 classical bits
qc.h(0)
qc.cx(0,1)
'''A measure() is added to the qasm simulator to get reasonable results in our measurements
   The first parameter is the list of qubits that will be measured 
   The second is a list of the classical bits where the result will be placed'''
qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])
backend = Aer.get_backend('qasm_simulator')
result = execute(qc, backend, shots=100).result()#shots=100
plot_histogram(result.get_counts())
```
